```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Did I really need to break this DFs down like this? 
#I feel like I am missing something...
df = pd.read_csv("cashew_data.csv")
ivc = df.loc[df['Country or Area'] == "CÃ´te d'Ivoire"]
gha = df.loc[df['Country or Area'] == "Ghana"]
burk = df.loc[df['Country or Area']== "Burkina Faso"]
ind = df.loc[df['Country or Area'] == "India"]
viet = df.loc[df['Country or Area'] == "Viet Nam"]
moz = df.loc[df['Country or Area']== 'Mozambique']
ng = df.loc[df['Country or Area']== 'Nigeria']
ben = df.loc[df['Country or Area']== 'Benin']
ivcexp = ivc.loc[ivc['Flow'] == 'Export']
vietexp = viet.loc[viet['Flow'] =='Export']
mozexp = moz.loc[moz['Flow'] =='Export']
indexp = ind.loc[ind['Flow']=='Export']
vietimp = viet.loc[viet['Flow']=='Import']
indimp = ind.loc[ind['Flow']=='Import']
indimp = ind.loc[ind['Flow']=='Import']
ngexp = ng.loc[ng['Flow']=='Export']
ghaexp = gha.loc[gha['Flow']=='Export']
burkexp = burk.loc[burk['Flow']=='Export']
benexp = ben.loc[ben['Flow']=='Export']
ivc_val = (ivcexp["Year"], ivcexp['Trade (USD)'])
#print(df)
```

```python
plt.figure(figsize=(15, 10))

plt.plot(ivcexp['Year'], ivcexp['Weight (kg)']/1000)
plt.ylabel('In Metric Tons')
plt.xlim(1998, 2016)
plt.title('Tons Exported - Cote DIvoire')
plt.xlabel('Year')
print("This shows the rapid growth in Cote D'Ivoire's Cashew Exports")
plt.show() 
print("This shows the rapid growth in Cote D'Ivoire's raw cashew exports")
print("Notice that cashew the cashew crop was barely 30,000 MT in 1997") 
print("Then the rate of growth increases from 2011 on ward.")
print('***!***Cote DIvoire = IVC***!***')
```

```python
print('This graph looks similar, but this is the value (in USD) of the exports')
plt.figure(figsize=(15, 10))

plt.plot(ivc_val[0], ivc_val[1]/1000000)
plt.xlim(1998, 2016)
plt.ylabel('USD (in millions)')
plt.xlabel('Year')
plt.title("IVC - Value of cashew exports in USD 1998-2015")
plt.show()
print("Notice how the jump in revenue lagged slightly behind the exports?")
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.plot(ivcexp['Year'], ivcexp['Quantity']/1000, color='green')
plt.xlim(1998, 2016)
plt.xlabel('Year')
plt.ylabel('in MT')
plt.title('Quantity')

plt.subplot(1, 2, 2)
plt.plot(ivcexp['Year'], ivcexp['Trade (USD)']/1000000)
plt.xlim(1998, 2016)
plt.xlabel ('Year')
plt.ylabel('USD in millions')
plt.title('Trade Value')

print('These graphs show the rise in quantity of tons (l) and total value of the quantity traded (r)')
plt.show()

print("That was cool, lets see those on the same chart!")
```

```python
plt.figure(figsize=(15, 10))
plt.plot(ivcexp['Year'], ivcexp['Quantity']/1000, color='green')
plt.plot(ivcexp['Year'], ivcexp['Trade (USD)']/1000)
plt.xlim(1998, 2016)
plt.ylabel('USD in 1000s(blue), Weight in MT(green)')
plt.xlabel('Year')
print("This is just to visually contrast the trend lines")
plt.show()
print("In 2011, you can tell that the price for a ton must have been close to USD 1000")
print("let's derive the price chart!")
```

```python
print('We can derive the average export price of a MT of IVC Cashew')
print('We do this by dividing total value by the total quantity')
plt.figure(figsize=(15,10))
plt.plot(ivcexp['Year'], ivcexp['Trade (USD)']/ivcexp['Quantity']*1000)
plt.xlim(1998, 2015)
plt.xlabel('Year')
plt.ylabel('USD/MT')
plt.title('Cashew Price - IVC 1998-2015')
plt.show()
print('Indeed you can see that price was around  USD 1000 in 2011')
```

```python
print("Let's see how IVC stacks up to the rest of Africa")
plt.figure(figsize=(15, 10))

plt.plot(ivcexp['Year'], ivcexp['Quantity']/1000, color='green')
plt.plot(ngexp['Year'], ngexp['Weight (kg)']/1000, color='blue')
plt.plot(mozexp['Year'], mozexp['Weight (kg)']/1000, color='purple')
plt.plot(ghaexp['Year'], ghaexp['Weight (kg)']/1000, color='red')
plt.plot(burkexp['Year'], burkexp['Weight (kg)']/1000, color='yellow')
plt.plot(benexp['Year'], benexp['Weight (kg)']/1000, color='pink')
plt.xlim(1998, 2016)
plt.show()
print('IVC=Green, Nigeria=Blue, Moza=Purple, Burkina=Yellow, Benin=Pink, Ghana=Red')
print("Wow!")
print("That's a lot of cashew! As African countries process very little cashew")
print("locally, one would wonder where all this cashew went...?")
print('The answer to that question occured half way around the globe')
print("and shows how IVC helped Viet Nam eat India's dinner...")
```

```python
print("So check this out, india imports raw cashew")
print("...and exports cashew kernels after the raw nuts are processed")
print('However, they also consume quite a bit domestically')
print("Check out how closely India's Imports tracks with IVC's exports")
plt.figure(figsize=(15,10))

plt.plot(indimp['Year'], indimp['Quantity']/1000, color='orange')
plt.plot(ivcexp['Year'], ivcexp['Quantity']/1000, color='green')
plt.xlim(1998, 2016)
plt.xlabel('Year')
plt.ylabel('Metric Tons')
plt.title('India Imports vs IVC Exports 1998-2015')
plt.show()

print('This shows india imports and exports')
print('However when processing cashew, the final product')
print('Is just 20%-22% of the original weight.')
print('I.e 1 kg of raw cashew gives you 220 grams or .220 kg')
plt.figure(figsize=(15,10))

plt.plot(indimp['Year'], indimp['Quantity']/1000, color='green')
plt.plot(indexp['Year'], indexp['Quantity']/1000 , color='red')
plt.xlim(1998, 2016)
plt.xlabel("Year")
plt.ylabel("Metric Tons")
plt.title("India - Imports(green) vs Exports(red)")
plt.show()

print("so to 'normalize' the data of out flow so that we can compare to inflow")
print("assume 22% output, then 1/.22 = 4.5454 (mlt)")
print("let's what happens when we multiply the outflow kg by our new mlt")
```

```python
plt.figure(figsize=(15,10))
plt.plot(indimp['Year'], indimp['Quantity']/1000, color='green')
plt.plot(indexp['Year'], indexp['Quantity']/1000 * 4.5454, color='red')
plt.xlabel("Year")
plt.ylabel("Metric Tons")
plt.title("India - Imports(green) vs Exports*scalor(red)")
plt.show()

print('Now, because India also produces raw cashew, it is beyond my scope to comment much')
```

```python
plt.figure(figsize=(15,10))

plt.subplot(1, 2, 1)
plt.plot(vietimp['Year'], vietimp['Quantity']/1000, color='green')
plt.plot(vietexp['Year'], vietexp['Quantity']/1000*4.5454, color='red')
plt.title('Vietnam - Imports(green) vs Exports(red)')
plt.xlabel('Year')
plt.ylim(1000, 1500000)
plt.ylabel('Metric Tons')

plt.subplot(1, 2, 2)
plt.plot(indimp['Year'], indimp['Quantity']/1000, color='green')
plt.plot(indexp['Year'], indexp['Quantity']/1000*4.5454, color='red')
plt.xlabel("Year")
plt.ylim(1000, 1500000)
plt.ylabel("Metric Tons")
plt.title("India - Imports(green) vs Exports(red)")

plt.show()

print("In Viet Nam, you can see that from 2000-2011")
print("growth in imports(green) was relatively stable")
print("Exports were also steady")
```

```python
print("in 2010-2011, it looks like Viet Nam was successfully processing 100% of it's local crop")
print("technological advances saw the price of processing get slashed, 1000s of factories sprang up")
print("Viet Nam now needed to import raw material to keep it's industry alive")
print("enter *IVC*, notice how after 2010 vietnams import and export increase reltive to each other?")
print("here is a chart of Viet nam and india imports vs IVC's exports")

plt.figure(figsize=(15,10))

plt.plot(ivcexp['Year'], ivcexp['Quantity']/1000, color='blue')
plt.plot(vietimp['Year'], vietimp['Quantity']/1000, color='green')
plt.plot(indimp["Year"], indimp['Quantity']/1000, color='red')
plt.ylim(10000, 1000000)
plt.ylabel ('Metric Tons')
plt.xlabel('Year')
plt.title("IVC EXP -Blue, INDIA IMP -red, Viet IMP - green")
plt.show()

print("Vietnam and India make up almost 90% of cashew processing")
print("IVC is the world's largest producer of raw cashew nuts.")
print("Conclusion - Without IVC, Viet Nam could not have over taken India in kernel exports")
```
